## 🔧 IMPLEMENTATION GUIDANCE: Key Patterns to Follow

### **Pattern 1: Safety-First Default (Paper Mode)**
All code should follow this template:
```python
async def execute_order(self, order: Order) -> OrderResult:
    """Execute order with paper trading default."""
    if not config.paper_mode and not self.has_valid_keys():
        raise ConfigError("Live trading requires paper_mode=false AND valid API keys")
    
    if config.paper_mode:
        return await self._paper_execute(order)
    
    # Real execution only reachable if both conditions above pass
    return await self._live_execute(order)
```

### **Pattern 2: Idempotency Wrapper**
```python
async def with_idempotency(client_id: str, operation):
    """Wrap any operation with idempotency guarantees."""
    # Check if already executed
    cached = await redis.get(f"idempotency:{client_id}")
    if cached:
        return json.loads(cached)
    
    # Execute once
    result = await operation()
    
    # Cache for 24 hours (prevents retries from double-execute)
    await redis.setex(
        f"idempotency:{client_id}",
        86400,
        json.dumps(result)
    )
    return result
```

### **Pattern 3: Circuit Breaker Decorator**
```python
@circuit_breaker(
    failure_threshold=5,
    recovery_timeout=60,
    expected_exception=ExchangeError
)
async def exchange_api_call(self, *args, **kwargs):
    """Auto-recover from exchange API failures."""
    # Your exchange call here
    pass
```

### **Pattern 4: Distributed Tracing**
```python
from opentelemetry import trace

tracer = trace.get_tracer(__name__)

```

### **Pattern 5: Observable Metrics**
```python
from prometheus_client import Histogram, Counter, Gauge

order_latency_ms = Histogram(
    'order_latency_ms',
    'Order placement latency',
    buckets=[1, 5, 10, 50, 100, 500, 1000, 5000]
```

````python
from pathlib import Path

repo = Path('/workspaces/CTO-TEST-AI-trading-Bot')

checks = [
    ('Idempotency', 'core/idempotency.py'),
    ('Order Manager', 'execution/order_manager.py'),
    ('Circuit Breaker', 'core/circuit_breaker.py'),
    ('Retry Logic', 'core/retry.py'),
    ('Distributed Tracing', 'monitoring/tracing.py'),
    ('Risk Manager', 'execution/risk_manager.py'),
    ('Smart Order Router', 'execution/smart_order_router.py'),
    ('Kraken Executor', 'execution/kraken_executor.py'),
    ('Fed Calendar', 'data_ingestion/fed_calendar.py'),
    ('Economic Releases', 'data_ingestion/economic_releases.py'),
    ('Macro Aggregator', 'data_ingestion/macro_aggregator.py'),
    ('Feature Engineering', 'engine/feature_engineering.py'),
    ('Model Trainer', 'engine/model_trainer.py'),
    ('Ensemble Scorer', 'engine/ensemble_scorer.py'),
    ('Dashboard API', 'interface/dashboard_api.py'),
    ('WebSocket Manager', 'interface/websocket_manager.py'),
    ('Orders Routes', 'interface/routes/orders.py'),
    ('Risk Routes', 'interface/routes/risk.py'),
    ('Rust Order Matcher', 'rust/order-matcher/src/lib.rs'),
    ('Rust Risk Engine', 'rust/risk-engine/src/lib.rs'),
    ('Rust Gateway', 'rust/gateway/src/lib.rs'),
]

placeholder_markers = (
    'placeholder',
    'stub',
    'not implemented',
    'todo',
    'would ',
)


def is_code_heavy(text: str) -> bool:
    code_indicators = ['def ', 'class ', 'async def ', 'impl ', 'pub fn ', '@router', 'return ']
    return sum(text.count(x) for x in code_indicators) >= 3


def detect_status(path: Path) -> tuple[str, str]:
    if not path.exists():
        return 'MISSING', 'file not found'

    text = path.read_text(encoding='utf-8', errors='ignore')
    lower = text.lower()

    marker_hits = sum(1 for m in placeholder_markers if m in lower)
    heavy = is_code_heavy(text)

    if marker_hits >= 2 and not heavy:
        return 'STUB/PARTIAL', f'placeholder markers={marker_hits}'
    if marker_hits >= 1 and '@router' in text:
        return 'PARTIAL', f'route placeholders markers={marker_hits}'
    if heavy:
        return 'IMPLEMENTED', 'substantial executable logic detected'
    return 'PARTIAL', 'exists but limited logic'


rows = []
for name, rel in checks:
    status, evidence = detect_status(repo / rel)
    rows.append((name, rel, status, evidence))

# Extra check: integration tests presence
integration_dir = repo / 'tests/integration'
integration_tests = sorted([p.name for p in integration_dir.glob('test_*.py')]) if integration_dir.exists() else []

print('REMAINING AUDIT - VERIFIED STATUS')
print('=' * 110)
print(f"{'Component':30s} {'Status':15s} {'Path':45s} Evidence")
print('-' * 110)
for name, rel, status, evidence in rows:
    print(f"{name:30s} {status:15s} {rel:45s} {evidence}")
print('=' * 110)

status_counts = {}
for _, _, s, _ in rows:
    status_counts[s] = status_counts.get(s, 0) + 1

print('\nSummary Counts:')
for k in sorted(status_counts.keys()):
    print(f'- {k}: {status_counts[k]}')

print('\nIntegration Test Coverage:')
if integration_tests:
    print(f'- Found {len(integration_tests)} integration test files:')
    for t in integration_tests:
        print(f'  - {t}')
else:
    print('- No integration test files found under tests/integration (blocker)')

print('\nTop Live-Deployment Blockers (current):')
print('1. API hardening (auth/rate-limit/CORS policy)')
print('2. Placeholder route handlers in interface/routes/* need backend wiring')
print('3. Full VaR/CVaR + stress scenarios still missing')
print('4. Kraken executor still minimal')
print('5. Integration test suite is absent')
````
